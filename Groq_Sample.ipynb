```python
import os
from groq import Groq
from dotenv import load_dotenv
load_dotenv()

groq_api_key = os.getenv("GROQ_API_KEY")
os.environ["LANGCHAIN_API_KEY"] = os.getenv('LANGCHAIN_API_KEY')
os.environ["LANGCHAIN_TRACING_V2"] = "true"
os.environ["LANGCHAIN_PROJECT"] = os.getenv("LANGCHAIN_PROJECT")
```

```python
from langchain_groq import ChatGroq
model = ChatGroq(model = "llama-3.3-70b-versatile", api_key = groq_api_key)
```

```python
from langchain_core.messages import HumanMessage, SystemMessage

messages = [
        SystemMessage(content= "You are a expert in elden ring and your job is to guide the players who have any questions about the game"),
        HumanMessage(content= "I am usign blood hound sword, but I can't recall the name of it")
]
```

```python
response = model.invoke(messages)
response.content
```

```python
#parsing output
from langchain_core.output_parsers import StrOutputParser
parser = StrOutputParser()
chain = model | parser

chain.invoke(messages)
```

```python
#prompt templates
from langchain_core.prompts import ChatPromptTemplate
generic_template = "You are a expert in elden ring and your job is to guide the players who have any questions about the game"
prompt = ChatPromptTemplate.from_messages(
    [('system', generic_template),
     ('user', "what do i do next, I have defeated {boss name}")]
)



chain = prompt | model | parser
```

```python
result = chain.invoke('morgot')
```
